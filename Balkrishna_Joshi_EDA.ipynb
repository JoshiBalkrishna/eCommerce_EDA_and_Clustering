# EDA on Customers Dataset

```python
# Extracting the year and month from SignupDate
customers_data['SignupYear'] = customers_data['SignupDate'].dt.year
customers_data['SignupMonth'] = customers_data['SignupDate'].dt.month


```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=customers_data, x='SignupYear', palette='viridis')
plt.title('Customer Signup Trends by Year')
plt.xlabel('Year')
plt.ylabel('Number of Signups')
plt.show()

```

```python
# Signup month distribution
signup_month_dist = customers_data['SignupDate'].dt.month.value_counts().sort_index()

plt.figure(figsize=(8, 6))
signup_month_dist.plot(kind='bar', color='skyblue')
plt.title('Signup Distribution by Month')
plt.xlabel('Month')
plt.ylabel('Number of Signups')
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)
plt.show()
```

```python
# Signup trends per region
region_year_signup = customers_data.groupby(['Region', 'SignupYear']).size().unstack()

plt.figure(figsize=(12, 6))
region_year_signup.T.plot(kind='line', marker='o')
plt.title('Signup Trends by Region Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Signups')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Plot region-wise distribution
region_distribution = customers_data['Region'].value_counts()

plt.figure(figsize=(8, 6))
region_distribution.plot(kind='bar', color='teal')
plt.title('Customer Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

# EDA on Products Dataset

```python
# Count of products in each category
category_counts = products_data['Category'].value_counts()

plt.figure(figsize=(8, 6))
category_counts.plot(kind='bar', color='orange')
plt.title('Product Distribution by Category')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.show()
```

```python
# Price statistics
print("Price Statistics:", products_data['Price'].describe())

plt.figure(figsize=(8, 6))
sns.histplot(products_data['Price'], kde=True, bins=20, color='purple')
plt.title('Price Distribution of Products')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Price vs. Category
plt.figure(figsize=(10, 6))
sns.boxplot(data=products_data, x='Category', y='Price', palette='muted')
plt.title('Price Distribution Across Categories')
plt.xlabel('Product Category')
plt.ylabel('Price (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Product Price Outliers
# Identify outliers based on price
q1 = products_data['Price'].quantile(0.25)
q3 = products_data['Price'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = products_data[(products_data['Price'] < lower_bound) | (products_data['Price'] > upper_bound)]
print("Number of Outliers in Product Prices:", outliers.shape[0])


outliers
```

# EDA on Transactions Dataset

```python
# Extract year and month from TransactionDate
transactions_data['TransactionYear'] = transactions_data['TransactionDate'].dt.year
transactions_data['TransactionMonth'] = transactions_data['TransactionDate'].dt.month


monthly_trend = transactions_data.groupby(['TransactionYear', 'TransactionMonth']).size().reset_index(name='TransactionCount')

```

```python
# Line plot for transaction trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_trend, x='TransactionMonth', y='TransactionCount', hue='TransactionYear', marker='o')
plt.title('Monthly Transaction Trends')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.legend(title='Year')
plt.show()
```

```python
##Top-Selling Products

# Calculate total quantity and revenue per product
top_products = transactions_data.groupby('ProductID').agg({
    'Quantity': 'sum',
    'TotalValue': 'sum'
}).sort_values(by='Quantity', ascending=False).head(10)

```

```python
# Plot top products by quantity sold
plt.figure(figsize=(10, 6))
sns.barplot(x=top_products['Quantity'], y=top_products.index, palette='coolwarm')
plt.title('Top-Selling Products by Quantity')
plt.xlabel('Quantity Sold')
plt.ylabel('Product ID')
plt.show()
```

```python
# Customer Purchase Frequency
purchase_frequency = transactions_data['CustomerID'].value_counts()

plt.figure(figsize=(10, 6))
sns.histplot(purchase_frequency, bins=30, kde=False, color='blue')
plt.title('Customer Purchase Frequency')
plt.xlabel('Number of Transactions')
plt.ylabel('Number of Customers')
plt.show()
```

```python
# Revenue Trends 
transactions_data['YearMonth'] = transactions_data['TransactionDate'].dt.to_period('M')
monthly_revenue = transactions_data.groupby('YearMonth')['TotalValue'].sum()

plt.figure(figsize=(12, 6))
monthly_revenue.plot(kind='line', marker='o', color='purple')
plt.title('Monthly Revenue Trends')
plt.xlabel('Year-Month')
plt.ylabel('Total Revenue (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Between Quantity and Price
plt.figure(figsize=(8, 6))
sns.scatterplot(data=transactions_data, x='Quantity', y='Price', alpha=0.7)
plt.title('Quantity vs. Product Price')
plt.xlabel('Quantity Purchased')
plt.ylabel('Product Price (USD)')
plt.show()
```

# Combined Dataset Analysis

```python
merged_data = transactions_data.merge(customers_data, on='CustomerID', how='left')
merged_data = merged_data.merge(products_data, on='ProductID', how='left')
```

```python
# Revenue contribution by category
category_revenue = merged_data.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
category_revenue.plot(kind='bar', color='coral')
plt.title('Revenue Contribution by Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Revenue by region
region_revenue = merged_data.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
region_revenue.plot(kind='bar', color='gold')
plt.title('Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Total Revenue (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Customer lifetime value
customer_ltv = merged_data.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False)

# Top 10 customers 
top_ltv = customer_ltv.head(10)

plt.figure(figsize=(10, 6))
top_ltv.plot(kind='bar', color='pink')
plt.title('Top 10 Customers by Lifetime Value')
plt.xlabel('Customer ID')
plt.ylabel('Lifetime Value (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pairplot for numerical columns (Quantity, Price, TotalValue)
sns.pairplot(transactions_data[['Quantity', 'Price', 'TotalValue']])
plt.show()
```

```python
# Create a pivot table for customer-product interactions
customer_product_pivot = transactions_data.pivot_table(index='CustomerID', columns='ProductID', values='Quantity', aggfunc='sum', fill_value=0)

# Plot heatmap for customer-product interactions
plt.figure(figsize=(12, 8))
sns.heatmap(customer_product_pivot, cmap='YlGnBu', cbar_kws={'label': 'Quantity Purchased'}, linewidths=0.5)
plt.title('Customer-Product Interaction Heatmap')
plt.show()
```
